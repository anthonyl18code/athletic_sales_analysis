```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
df_2020 = pd.read_csv('athletic_sales_2020.csv')
df_2021 = pd.read_csv('athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
print(df_2020)
```

```python
# Display the 2021 sales DataFrame
print(df_2021)
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
df_2020 = pd.read_csv('athletic_sales_2020.csv')
print("Data types of the 2020 Sales DataFrame:")
print(df_2020.dtypes)
```

```python
# Check the 2021 sales data types.
df_2020 = pd.read_csv('athletic_sales_2020.csv')
print("Data types of the 2020 Sales DataFrame:")
print(df_2020.dtypes)
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_df = pd.concat([df_2020, df_2021], axis=0).reset_index(drop=True)
print("Combined DataFrame:")
print(combined_df.head())
```

```python
# Check if any values are null.
print("\nCheck for null values in the combined DataFrame:")
print(combined_df.isnull().sum())
```

```python
# Check the data type of each column
print("\nData types of the combined DataFrame:")
print(combined_df.dtypes)
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_df['invoice_date'] = pd.to_datetime(combined_df['invoice_date'], format='%Y-%m-%d')

print("\nData types after converting 'invoice_date':")
print(combined_df.dtypes)
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_df['invoice_date'] = pd.to_datetime(combined_df['invoice_date'], format='%Y-%m-%d')
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".


# Show the top 5 results.
region_products = combined_df.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()
region_products = region_products.rename(columns={'units_sold': 'Total_Products_Sold'})
region_products = region_products.sort_values(by='Total_Products_Sold', ascending=False)

print("Top 5 regions by Total Products Sold:")
print(region_products.head(5))
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.


# Rename the "units_sold" column to "Total_Products_Sold"


# Show the top 5 results.
pivot_result = combined_df.pivot_table(
    index=['region', 'state', 'city'],  
    values='units_sold',               
    aggfunc='sum'                      
).reset_index()

pivot_result = pivot_result.rename(columns={'units_sold': 'Total_Products_Sold'})
pivot_result = pivot_result.sort_values(by='Total_Products_Sold', ascending=False)

print("Top 5 regions by Total Products Sold using pivot_table:")
print(pivot_result.head(5))
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.
region_sales = combined_df.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()
region_sales = region_sales.sort_values(by='Total_Sales', ascending=False)

print("Top 5 regions by Total Sales:")
print(region_sales.head(5))
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.


# Optional: Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.
pivot_sales = combined_df.pivot_table(
    index=['region', 'state', 'city'],
    values='total_sales',              
    aggfunc='sum'                      
).reset_index()

pivot_sales = pivot_sales.rename(columns={'total_sales': 'Total_Sales'})
pivot_sales = pivot_sales.sort_values(by='Total_Sales', ascending=False)

print("Top 5 regions by Total Sales using pivot_table:")
print(pivot_sales.head(5))
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.
retailer_sales = combined_df.groupby(['retailer', 'region', 'state', 'city'])['total_sales'].sum().reset_index()
retailer_sales = retailer_sales.rename(columns={'total_sales': 'Total_Sales'})
retailer_sales = retailer_sales.sort_values(by='Total_Sales', ascending=False)

print("Top 5 retailers by Total Sales:")
print(retailer_sales.head(5))
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.


# Optional: Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.
pivot_retailer_sales = combined_df.pivot_table(
    index=['retailer', 'region', 'state', 'city'],
    values='total_sales',                          
    aggfunc='sum'                                  
).reset_index()

pivot_retailer_sales = pivot_retailer_sales.rename(columns={'total_sales': 'Total_Sales'})
pivot_retailer_sales = pivot_retailer_sales.sort_values(by='Total_Sales', ascending=False)

print("Top 5 retailers by Total Sales using pivot_table:")
print(pivot_retailer_sales.head(5))
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_footwear_sales = combined_df[combined_df['product'].str.contains("Women's Athletic Footwear", case=False)]

print("Filtered data for Women's Athletic Footwear:")
print(womens_footwear_sales.head())
print(f"\nNumber of rows in filtered data: {len(womens_footwear_sales)}")
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_sales = combined_df[combined_df['product'].str.contains("Women's Athletic Footwear", case=False)]
womens_units_sold = womens_footwear_sales.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().reset_index()
womens_units_sold = womens_units_sold.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})
womens_units_sold = womens_units_sold.sort_values(by='Womens_Footwear_Units_Sold', ascending=False)

# Show the top 5 results.
print("Top 5 retailers by Women's Footwear Units Sold:")
print(womens_units_sold.head(5))
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_footwear_sales = combined_df[combined_df['product'].str.contains("Women's Athletic Footwear", case=False)]
pivot_womens_units = womens_footwear_sales.pivot_table(
    index=['retailer', 'region', 'state', 'city'],
    values='units_sold',                           
    aggfunc='sum'                                  
).reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
pivot_womens_units = pivot_womens_units.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})
pivot_womens_units = pivot_womens_units.sort_values(by='Womens_Footwear_Units_Sold', ascending=False)


# Show the top 5 results.
print("Top 5 retailers by Women's Footwear Units Sold using pivot_table:")
print(pivot_womens_units.head(5))
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
womens_footwear_sales = combined_df[combined_df['product'].str.contains("Women's Athletic Footwear", case=False)]
pivot_daily_sales = womens_footwear_sales.pivot_table(
    index='invoice_date',         
    values='total_sales',         
    aggfunc='sum'                 
).reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_daily_sales = pivot_daily_sales.rename(columns={'total_sales': 'Total_Sales'})
pivot_daily_sales = pivot_daily_sales.sort_values(by='Total_Sales', ascending=False)


# Show the table.
print("Daily Women's Athletic Footwear Sales:")
print(pivot_daily_sales)

top_day = pivot_daily_sales.iloc[0]
print(f"\nThe day with the most sales is {top_day['invoice_date']} with Total Sales of {top_day['Total_Sales']}.")
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
womens_footwear_sales['invoice_date'] = pd.to_datetime(womens_footwear_sales['invoice_date'])
womens_footwear_sales.set_index('invoice_date', inplace=True)
daily_sales = womens_footwear_sales['total_sales'].resample('D').sum().reset_index()

# Sort the resampled pivot table in descending order on "Total Sales".
daily_sales = daily_sales.rename(columns={'total_sales': 'Total_Sales'})
daily_sales = daily_sales.sort_values(by='Total_Sales', ascending=False)

print("Daily Women's Athletic Footwear Sales (Resampled):")
print(daily_sales.head(10))
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
womens_footwear_sales['invoice_date'] = pd.to_datetime(womens_footwear_sales['invoice_date'])
womens_footwear_sales.set_index('invoice_date', inplace=True)
weekly_sales = womens_footwear_sales['total_sales'].resample('W').sum().reset_index()

# Sort the resampled pivot table in descending order on "Total Sales".
weekly_sales = weekly_sales.rename(columns={'total_sales': 'Total_Sales'})
weekly_sales = weekly_sales.sort_values(by='Total_Sales', ascending=False)

print("Weekly Women's Athletic Footwear Sales (Resampled):")
print(weekly_sales.head(10))
```
